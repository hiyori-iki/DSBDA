5. Data formatting and normalization: 

Replacing NaN values of calories with mean

```python
x=df['Calories'].mean()
df['Calories'].fillna(x,inplace= True)
df
```

Changing NaN to NaT

```python
df['Date']=pd.to_datetime(df['Date'])
df
```

Removing negative sign of calories by absolute function

```python
df['Calories']=df['Calories'].abs()
df
```

Dropping 22nd row with NaT value of Date

```python
df.dropna(subset=['Date'], inplace=True)
df
```

Replacing outliers i.e.450 in 7th row with mean value.

```python
x=df['Duration'].mean()
df.loc[7,'Duration']=x
df
```
